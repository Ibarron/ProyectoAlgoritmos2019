```python
Datos = loadtxt("2019_05_14-17_16_25_Summary.csv", delimiter=",",skiprows=1,usecols=range(1,34))
print(len(Datos))
print(Datos.shape)
print(Datos)
```

```python
Acc=Datos[:,5]; #Toma todos los valores especificos de la columna 5
plt.plot(Acc,'g')
plt.grid(ls=":")
plt.xlabel('t (s)')
plt.ylabel('a (g)')
plt.show()
```

```python
tMin=307
tMax=1300
plt.plot(Acc[tMin:tMax],'g')
plt.grid(ls=":")
plt.xlabel('t (s)')
plt.ylabel('a (g)')
plt.show()
```

```python
HR=Datos[:,0];
plt.plot(HR,'r')
plt.ylim(50,180)
plt.grid(ls=":")
plt.axvspan(tMin,tMax,facecolor='g',alpha=.25)
plt.xlabel('t (s)')
plt.ylabel('HR (bpm)')
plt.show()
```

```python
# Frecuencia respiratoria (breathing rate BR) 
# se mide a partir de los intervalos de tiempo entre 2 inspiraciones o expiraciones sucesivas
# se mide en ciclos por minuto (cycles per minute cpm)
BR=Datos[:,1];
plt.plot(BR,'grey')
plt.grid(ls=":")
plt.axvspan(tMin,tMax,facecolor='g',alpha=0.25)
plt.xlabel('t (s)')
plt.ylabel('BR (cpm)')
plt.show()
```

```python
# La amplitud respiratoria (breathing ampltitude BA) se estima a partir del estira-afloja del cintur√≥n
# BA se mide en unidades arbitrarias (arbitrary units a.u.)
BA=Datos[:,8];
plt.plot(BA,'grey')
plt.ylim(4000,40000)
plt.grid(ls=":")
plt.axvspan(tMin,tMax,facecolor='g',alpha=0.25)
plt.xlabel('t (s)')
plt.ylabel('BA (a.u.)')
plt.show()
```

```python
Accel = np.loadtxt("2019_05_14-17_16_25_Accel.csv", delimiter=",",skiprows=1,usecols=range(1,4))
print(len(Accel))
print(Accel.shape)
```

```python
# Seleccionar fragmento de la caminata del archivo de resumen, tomando en cuenta que la frecuencia de muestreo aqu√≠ es 100x mayor
aZ=Accel[100*tMin:100*tMax,0] 
aY=Accel[100*tMin:100*tMax,1]
aX=Accel[100*tMin:100*tMax,2]
```

```python
az=(aZ-2048)/83
ay=(aY-2048)/83
ax=(aX-2048)/83
```

```python
ayy=ay-mean(ay)
azz=az-mean(az)
axx=ax-mean(ax)
a=sqrt(axx**2+ayy**2+azz**2)
```

```python
aYY=ayy.shape[0]
aXX= axx.shape[0]
aZZ= azz.shape[0]
```

```python
ayyf=(2.0/aYY*abs(fft(ayy))**2)
plt.clf()
plt.plot(aYY)

plt.clf()
plt.plot(ayyf[0:aYY//2])#muestra todas las frecuencias
plt.grid()
plt.title ("Eje Y")
plt.xlabel('f')
plt.ylabel('P(f)')
xlim(0,10000)
plt.show()


axxf=(2.0/aXX*abs(fft(axx))**2)
plt.clf()
plt.plot(aXX)

plt.clf()
plt.plot(axxf[0:aXX//2])#muestra todas las frecuencias
plt.grid()
plt.title ("Eje X")
plt.xlabel('f')
plt.ylabel('P(f)')
xlim(0,10000)
plt.show()


azzf=(2.0/aZZ*abs(fft(azz))**2)
plt.clf()
plt.plot(aZZ)

plt.clf()
plt.plot(azzf[0:aZZ//2])#muestra todas las frecuencias
plt.grid()
plt.title ("Eje Z")
plt.xlabel('f')
plt.ylabel('P(f)')
xlim(0,10000)
plt.show()
```
